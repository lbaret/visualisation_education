# Test t-student sur les résultats par sexe

L'**objectif** est de savoir si les distributions entre les résultats pour un homme sont vraiment différents de ceux des femmes.  
Nous avons de la chance, nous savons que les distributions semblent suivre une loi normale, donc le test de Student va être pertinent dans notre cas.  
Cepdendant, vérifions que nous avons une distribution assimilable à une distribution normale quand nous séparons par sexe.

```python
gender_gr = df.groupby('gender')

# Split
df_male = gender_gr.get_group('male')
df_female = gender_gr.get_group('female')
```

## Vérification pour les maths

```python
plt.figure(figsize=(20, 10))

x = np.linspace(0, 100, 1001)

plt.hist(df_male['math score'], density=True, alpha=0.5, label='Homme', color='tab:blue', bins=20)
plt.plot(x, norm.pdf(x, loc=df_male['math score'].mean(), scale=df_male['math score'].std()), color='blue')

plt.hist(df_female['math score'], density=True, alpha=0.5, label='Femme', color='tab:orange', bins=20)
plt.plot(x, norm.pdf(x, loc=df_female['math score'].mean(), scale=df_female['math score'].std()), color='orange')
plt.legend()

plt.show()
```

Tout est nickel du côté des maths, quelque soit le sexe, la distribution est assimilable à une distribution normale. 😎

## Vérification pour la lecture

```python
plt.figure(figsize=(20, 10))

x = np.linspace(0, 100, 1001)

plt.hist(df_male['reading score'], density=True, alpha=0.5, label='Homme', color='tab:blue', bins=20)
plt.plot(x, norm.pdf(x, loc=df_male['reading score'].mean(), scale=df_male['reading score'].std()), color='blue')

plt.hist(df_female['reading score'], density=True, alpha=0.5, label='Femme', color='tab:orange', bins=20)
plt.plot(x, norm.pdf(x, loc=df_female['reading score'].mean(), scale=df_female['reading score'].std()), color='orange')
plt.legend()

plt.show()
```

Pour la lecture également. ✅

## Vérification écriture

```python
plt.figure(figsize=(20, 10))

x = np.linspace(0, 100, 1001)

plt.hist(df_male['writing score'], density=True, alpha=0.5, label='Homme', color='tab:blue', bins=20)
plt.plot(x, norm.pdf(x, loc=df_male['writing score'].mean(), scale=df_male['writing score'].std()), color='blue')

plt.hist(df_female['writing score'], density=True, alpha=0.5, label='Femme', color='tab:orange', bins=20)
plt.plot(x, norm.pdf(x, loc=df_female['writing score'].mean(), scale=df_female['writing score'].std()), color='orange')
plt.legend()

plt.show()
```

Pour les hommes c'est légérement ambigüe, mais considérons que ça passe. 😅

**Conclusion**

Maintenant que nous avons contrôler nos distributions, nous pouvons procéder sans crainte au t-test de *Student*. 🤩

**À savoir**

D'après la documentation de *scipy* sur le test de *Student* indépendant :  
> *We can use this test, if we observe two independent samples from the same or different population, e.g. exam scores of boys and girls or of two ethnic groups. The test measures whether the average (expected) value differs significantly across samples. If we observe a large p-value, for example larger than 0.05 or 0.1, then we cannot reject the null hypothesis of identical average scores. If the p-value is smaller than the threshold, e.g. 1%, 5% or 10%, then we reject the null hypothesis of equal averages.*

## t-test maths

```python
ttest_ind(df_male['math score'], df_female['math score']).pvalue
```

Le résultat est intéressant : $p-value < 0.05$ donc nous pouvons rejeter l'hypothèse nulle $H_0$ qui stipule que les deux moyennes des distributions sont identiques.

**Conclusion**

Les hommes seraient meilleurs que les femmes en mathématiques. 👨‍🎓

## t-test lecture

```python
ttest_ind(df_male['reading score'], df_female['reading score']).pvalue
```

Nous obtenons une conclusion similaire qu'à la situation précédente. Nous rejetons l'hypothèse nulle $H_0$.

**Conclusion**

Les femmes seraient meilleurs que les hommes en lecture. 👩‍🎓

## t-test écriture

```python
ttest_ind(df_male['writing score'], df_female['writing score']).pvalue
```

Nous obtenons une conclusion similaire qu'aux situations précédentes. Nous rejetons l'hypothèse nulle $H_0$.

**Conclusion**

Les femmes seraient meilleurs que les hommes en écriture. 👩‍🎓
